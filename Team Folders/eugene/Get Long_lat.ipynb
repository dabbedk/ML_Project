```python
import numpy as np
import pandas as pd
%matplotlib inline 
pd.options.display.max_columns = 50
import warnings
warnings.filterwarnings('ignore')
import geopy
from geopy import Nominatim
from geopy.extra.rate_limiter import RateLimiter

housing = pd.read_csv('./data/Ames_Housing_Price_Data.csv', index_col=0)
address = pd.read_csv('./data/Ames_Real_Estate_Data.csv', index_col=0)
housing.shape
```

```python
address = address['Prop_Addr']
```

```python
address
```

```python
housing = housing.merge(address, how='left', left_on='PID', right_index=True)
```

```python
housing.shape
```

```python
street_addresses = housing['Prop_Addr']
street_addresses
```

```python
import geopy
from geopy import Nominatim
from geopy.extra.rate_limiter import RateLimiter

locator = Nominatim(user_agent="myGeocoder")
geocode = RateLimiter(locator.geocode, min_delay_seconds=1.5)
```

```python
locations = [geocode(addr+', Ames, USA') if type(addr)==str else None for addr in street_addresses]
```

```python
locations[2].address
```

```python
locations
```

```python
longitude=[]
latitude=[]
addy=[]
for i in locations:
    try:
        longitude.append(i.longitude)
    except AttributeError:
        longitude.append("none")
    try:
        latitude.append(i.latitude)
    except AttributeError:
        latitude.append("none")
    try:
        addy.append(i.address)
    except AttributeError:
        addy.append("none")
```

```python
housing.drop_duplicates()
```

```python
housing['longitude']=longitude
housing['latitude']=latitude
housing['addy']=addy
housing.drop_duplicates()
```

```python
housing.to_csv('housing_coords.csv')
```
